```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LinearRegression 
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

data = load_iris(as_frame=True)
df = data.frame
df.head()
```

```python
X = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
y = df['target']
```

### Cross Validation
Cross-validation (CV) is a technique used to evaluate how well a machine learning model generalizes to unseen data.

It helps prevent:
- Overfitting
- Underfitting
- Data leakage

Why Do We Need Cross-Validation?

If you train on training data and test once on test data, you get one score.
- But that score may be:
- lucky
- unlucky
- not stable
- dependent on data split

â›” A single train/test split is not reliable.
Cross-validation gives a more stable, trustworthy performance estimate.

#### K-Fold Cross-Validation
ðŸ“Œ How it works:
1. Split dataset into K equal parts (folds)
2. Train on Kâˆ’1 folds
3. Test on the remaining fold
4. Repeat until every fold has been used as test
5. Average the scores

Example: K=5
Fold1  Fold2  Fold3  Fold4  Fold5

Step1: Train on 2,3,4,5 | Test on 1  
Step2: Train on 1,3,4,5 | Test on 2  
Step3: Train on 1,2,4,5 | Test on 3  
Step4: Train on 1,2,3,5 | Test on 4  
Step5: Train on 1,2,3,4 | Test on 5

Final score = average of 5 test scores

```python
# Instead of spliting data pass all data and it will divide data into 5 different folds
# Cross validation with fold 5 
lin_model = LinearRegression()
lin_foldScores = cross_validate(lin_model, X, y, cv=5, scoring='r2')
print("Score of each fold:", lin_foldScores['test_score'])
print("Linear model average of folds:", np.mean(lin_foldScores['test_score'])*100)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
# Cross validation with fold 5
knn_foldScores = cross_validate(knn_model, X, y, cv=5, scoring='accuracy')
print("Score of each fold:", knn_foldScores['test_score'])
print("KNN model average of folds:", np.mean(knn_foldScores['test_score'])*100)
```
