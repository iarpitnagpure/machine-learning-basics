```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error

# Read the data from CSV file
dataSet = pd.read_csv('../Study_Hours.csv')
dataSet.head()
```

```python
# Features (X) and target (y) 
X = dataSet[['Number of Hours Studied']]
y = dataSet['Marks Scored']

# Create and train the model using .fit() method
model = LinearRegression()
model.fit(X, y)

# Predict the values X values based on learning
y_pred = model.predict(X)
print("Predicted Values:", y_pred)

# Evaluate the Linear Regression performance
mae = mean_absolute_error(y, y_pred)
mse = mean_absolute_error(y, y_pred)
rmse = root_mean_squared_error(y, y_pred)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)

# Take user input for prediction
hourStudied = float(input("Enter hours of studied"))
# Create a DataFrame for the new input prediction since model.predict() expects a 2D array-like structure or pass a DataFrame
predictedScore = model.predict(pd.DataFrame({'Number of Hours Studied': [hourStudied]}))        # Entered 5 hours
# Output the prediction
print("Predicted Score:", predictedScore)
```
