```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score

dataSet = pd.read_csv('../Students_Performance_Dataset.csv')
dataSet.head()

# Features (X) and target (y) 
X = dataSet[['Study_Hours_per_Week']]
y = dataSet['Total_Score']

# Create and train the model using .fit() method
model = LinearRegression()
model.fit(X, y)

# Predict the values X values based on learning

y_pred = model.predict(X)
print("Y Prediction:", y_pred)

mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = root_mean_squared_error(y, y_pred)
r2Score = r2_score(y, y_pred)
print("MAE:", round(mae, 2))
print("MSE:", round(mae, 2))
print("RMSE:", round(rmse, 2))
print("R2Score:", round(r2Score, 4))
```

```python
# Plot histogram to check score (y) are evenly speard or not.
plt.figure(figsize=(10, 6))                                         # Plot size (10 x 6)
plt.hist(y, bins=30, color='lightblue', edgecolor='black')
plt.xlabel("Final score")
plt.ylabel("Number of Student")
plt.title("Distribution of final score")
plt.grid(True)
plt.show()
```

```python
# Plot histogram to check score are evenly speard or not.
plt.figure(figsize=(10, 6))                                         # Plot size (10 x 6)
plt.scatter(X, y, color='blue', label='Actual Score')
plt.plot(X, y_pred, color='red', label='Predicted score regression line')
plt.xlabel("Study hours per week")
plt.ylabel("Final Output")
plt.title("Model Prediction Score vs Actual Score")
plt.grid(True)
plt.show()
```
