```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

df = pd.DataFrame({
    'Hours_Studied': [1, 2, 3, 4, 5, 6, 7],
    'Passed': [0, 0, 0, 1, 1, 1, 1]
})
df
```

### Logistic Regression
Logistic Regression is a supervised learning algorithm used for classification, not regression (despite the name).
It is used when the output (target) is categorical, such as:
- Yes / No
- Spam / Not spam
- Disease / No disease
- 0 / 1

#### Key Idea
Logistic Regression does not predict values directly.
Instead, it predicts the probability that a given input belongs to a class.
t uses the Sigmoid Function to convert outputs into probability:

Where:
- Output is always between 0 and 1
- If probability > 0.5 → Class = 1
- If probability ≤ 0.5 → Class = 0

When to Use Logistic Regression:
| Type of Prediction                                 | Logistic Regression |
| -------------------------------------------------- | ------------------- |
| Predicting numeric continuous values (like prices) | ❌ No                |
| Predicting categories (Yes/No, 0/1)                | ✅ Yes               |


```python
# Features (X) and target (y)
X = df[['Hours_Studied']]
y = df['Passed']

# Create and train the model using .fit() method
model = LogisticRegression()
model.fit(X, y)

# Take user input for prediction
userInput = float(input("Enter number of hours studied: "))

# Create a DataFrame for the new input prediction since model.predict() expects a 2D array-like structure or pass a DataFrame
newDataFrame = pd.DataFrame({'Hours_Studied': [userInput]})
y_pred = model.predict(newDataFrame)

# Output the prediction
print("Predicted Exam Passing Result:", y_pred[0])                  # Input = 4, Based on that return 1 (Passed)
```
